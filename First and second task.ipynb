```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
from pandas_profiling import ProfileReport
```

## Data import and distribution plot

```python
train_df = pd.read_csv('https://raw.githubusercontent.com/Serfentum/bf_course/master/14.pandas/train.csv')
pos = train_df['pos']
a = train_df['A_fraction']
t = train_df['T_fraction']
g = train_df['G_fraction']
c = train_df['C_fraction']
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)

plt.title('Nucleotide distribution')
p1 = plt.bar(pos, a, color='r', label = 'A')
p2 = plt.bar(pos, t, bottom=a, color='b', label = 'T')
p3 = plt.bar(pos, g, bottom=np.array(a)+np.array(t), color='g', label = 'G')
p4 = plt.bar(pos, c, bottom=np.array(a)+np.array(t)+np.array(g), color='c', label = 'C')
plt.xlabel('Nucleotide position')
plt.ylabel('Nucleotide frequency')
plt.grid(axis = 'y')
plt.legend()
plt.show()
```

## Export of dataset part as train_part.csv

```python
mean_matches = train_df['matches'].mean()
train_df_part = train_df.query("matches > @mean_matches")[['pos', 'reads_all', 'mismatches', 'deletions', 'insertions']]
train_df_part.to_csv("train_part.csv")
```
